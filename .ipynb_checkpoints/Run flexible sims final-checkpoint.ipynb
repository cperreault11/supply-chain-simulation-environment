```python
from scse.main import miniSCOTnotebook
import numpy as np
import pandas as pd
import sys
```

```python
m = miniSCOTnotebook()
```

```python
run_params={
    'time_horizon':730,
    'extra_power': 2,
    'flexible_capacity': 2500000,
    'static_capacity': 5500000,
}
```

```python
renewable_scales = np.linspace(.1,.7,13)
```

```python
renewable_scales = [round(x,2) for x in renewable_scales]
```

```python
num_runs = 10
path = "./sims/renewable_scales/"
market_demand = 7000000
```

```python
for s in renewable_scales:
    for i in range(num_runs):
        sys.stdout = open(path + "renewable_{}_{}".format(s,i),'w')
        run_params['renewable_scale'] = market_demand * s
        m.start(**run_params)
        m.run()
```
